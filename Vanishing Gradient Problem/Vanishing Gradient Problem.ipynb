```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Creating a Classification dataset

from sklearn.datasets import make_moons    # make_moons generates a 2D binary classification dataset where the points form two crescent moon shapes.

x, y = make_moons(n_samples = 250, noise = 0.13, random_state = 42)
print(x[:5])
print(y[:5])
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)         # x[:, 0]: Feature-1 v/s x[:, 1]: Feature-2. & c=y tells color of each point will depend on the value in y.
plt.show()
```

```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
```

```python
# Building Neural n/w model

model = Sequential()

model.add(Dense(10, activation = 'sigmoid', input_dim = 2))    # I/P layer
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(10, activation = 'sigmoid'))
model.add(Dense(1, activation = 'sigmoid'))                    # O/P layer
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Paramters of 1st layer

model.get_weights()[0]
# get_weights() returns a list of NumPy arrays. Each array represents the weights or biases of a layer in your model.
```

```python
old_weights = model.get_weights()[0]  # Storing these weight to further compare
```

```python
# Split the dataset for training

x_train, x_test, Y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 42)
```

```python
# Training the model on the dataset

history = model.fit(x_train, Y_train, epochs = 20)
```

```python
# Updated values of paramters in 1 st layer

new_weights = model.get_weights()[0]
print('Paramters Before training:\n', old_weights, '\n\nUpdated paramters:\n', new_weights)

print('\nPercentage of change in parameters of 1st layer:\n', 100*(new_weights - old_weights)/old_weights)
```

### As you can see, the parameters are nearly the same after training, which means that the Vanishing Gradient Problem has occurred.

```python
# Plotting the weights

plt.scatter(old_weights, new_weights)
plt.show()
```

```python
# Plotting the loss

plt.ylim(0, 1)
plt.plot(history.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

## Evidence of Vanishing Gradient Problem

#### To analyze the training behavior of the model, we plotted the previous and updated weight values during training. The resulting graph showed a near-straight line, indicating minimal change in the weights across iterations. Additionally, the loss remained almost constant over time.

#### This behavior suggests that the gradients being computed during backpropagation are extremely small — a hallmark of the **vanishing gradient problem**. As a result, the model fails to update its parameters effectively, leading to stalled learning and no significant improvement in loss.
