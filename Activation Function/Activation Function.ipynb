```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating x-axis values

x = np.linspace(-10, 10, 400)     # Creates an array of 400 equally spaced values b/w -10 and 10
print(x[0:10])
```

## 1) Sigmoid Function

```python
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)
```

```python
# Generating y-axis values for corresponding x-axis values
y = sigmoid(x)
y_derivative = sigmoid_derivative(x)
print(y[0:5],'\n', y_derivative[0:5])
```

```python
# Plotting 

plt.plot(x, y, label = 'Sigmoid')
plt.plot(x, y_derivative, label = 'Sigmoid\'')

plt.title('Sigmoid Function')
plt.grid(True)
plt.legend()
plt.show()
```

## 2) Tanh Function

```python
# Tanh function
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

# Derivative
def tanh_derivative(x):
    val = tanh(x)
    return 1 - np.square(val)
```

```python
# Generating y-axis values for corresponding x-axis values
y = tanh(x)
y_derivative = tanh_derivative(x)

# Plotting 
plt.plot(x, y, label = 'Tanh')
plt.plot(x, y_derivative, label = 'Tanh\'')

plt.axhline(0, color='black', linewidth=0.5)  # highlight y-axis 
plt.axvline(0, color='black', linewidth=0.5)  # highlight x-axis

plt.title('Tanh Function')
plt.grid(True)
plt.legend()
plt.show()
```

## 3) ReLU Function

```python
# ReLU
def relu(x):
    return np.maximum(0, x)

# Derivative
def relu_derivative(x):
    return np.where(x > 0, 1, 0)
```

```python
x = np.linspace(-5, 5, 400)    # Taking x to be of array of 400 equally spaced points b/w -5 and 5, so that the graph seems more interpretable 

# Generating y-axis values for corresponding x-axis values
y = relu(x)
y_derivative = relu_derivative(x)

# Plotting 
plt.plot(x, y, label = 'ReLU')
plt.plot(x, y_derivative, label = 'ReLU\'')

plt.title('ReLU Function')
plt.grid(True)
plt.legend()
plt.show()
```

## 4) Parametric ReLU

```python
# Parametric relu
def para_relu(x, alpha):
    return np.where(x > 0, x, alpha*x)

def para_derivative(x, alpha):
    return np.where(x > 0, 1, alpha)
```

```python
# Generating y-axis values for corresponding x-axis values
y = para_relu(x, 0.5)
y_derivative = para_derivative(x, 0.5)

# Plotting 
plt.plot(x, y, label = 'Parametric ReLU')
plt.plot(x, y_derivative, label = 'Parametric ReLU\'')

plt.axhline(0, color='black', linewidth=0.5)  # highlight y-axis 
plt.axvline(0, color='black', linewidth=0.5)  # highlight x-axis

plt.title('Parametric ReLU Function')
plt.grid(True)
plt.legend()
plt.show()
```

Parametric ReLU is a variation of ReLU where the negative slope (`α`) is a learnable parameter rather than a fixed constant. This allows the network to adaptively learn the best value of `α` during training.

> When `α` is set to a small fixed value like **0.01**, Parametric ReLU behaves like **Leaky ReLU**.

## 5) Leaky ReLU

```python
# Generating y-axis values for corresponding x-axis values
y = para_relu(x, 0.01)
y_derivative = para_derivative(x, 0.01)

# Plotting 
plt.plot(x, y, label = 'Leaky ReLU')
plt.plot(x, y_derivative, label = 'Leaky ReLU\'')

plt.axhline(0, color='black', linewidth=0.5)  # highlight y-axis 
plt.axvline(0, color='black', linewidth=0.5)  # highlight x-axis

plt.title('Leaky ReLU Function')
plt.grid(True)
plt.legend()
plt.show()
```

## 6) Elu ReLU

```python
# Elu Relu
def elu(x, alpha):
    #if x <= 0:
     #   val = alpha*(np.exp(x) - 1)
    return np.where(x > 0, x, alpha*(np.exp(x) - 1))

# Derivative
def elu_derivative(x, alpha):
    c = alpha
    val = elu(x, c)
    return np.where(x > 0, 1, elu(x, c) + alpha)
```

```python
# Generating y-axis values for corresponding x-axis values
y = elu(x, 1)
y_derivative = para_derivative(x, 1)

# Plotting 
plt.plot(x, y, label = 'ELU')
plt.plot(x, y_derivative, label = 'ELU\'')

plt.axhline(0, color='black', linewidth=0.5)  # highlight y-axis 
plt.axvline(0, color='black', linewidth=0.5)  # highlight x-axis

plt.title('ELU Function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-10, 10, 1000)

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

# Plot all on the same figure
plt.figure(figsize=(10, 6))

plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, leaky_relu(x), label='Leaky ReLU')

# Add labels, title, and legend
plt.title("Activation Functions Comparison")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.show()
```
